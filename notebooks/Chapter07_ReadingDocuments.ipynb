```python
from urllib.request import urlopen
from io import StringIO
import csv

text = (
    urlopen("http://pythonscraping.com/files/MontyPythonAlbums.csv")
    .read()
    .decode("ascii", "ignore")
)
csv_data = csv.reader(StringIO(text))

for index, row in enumerate(csv_data):
    # ヘッダーはスキップ
    if index == 0:
        continue

    album_name, released_year = row[0], row[1]

    print(f"The album '{album_name}' was released in {released_year}")
```

```python
# PDF
from urllib.request import urlopen
from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO, BytesIO
from typing import IO


def read_pdf(pdf: IO[bytes]) -> str:
    resource_manager = PDFResourceManager()
    pdf_text_io = StringIO()
    layout_params = LAParams()

    try:
        converter = TextConverter(
            resource_manager,
            pdf_text_io,
            laparams=layout_params,
        )

        try:
            process_pdf(resource_manager, converter, pdf)
        finally:
            converter.close()

        content = pdf_text_io.getvalue()
        return content
    finally:
        pdf_text_io.close()
```

```python
pdf_url = "http://pythonscraping.com/pages/warandpeace/chapter1.pdf"

with urlopen(pdf_url) as pdf_file:
    pdf_text = read_pdf(BytesIO(pdf_file.read()))

print(pdf_text)
```

```python
from zipfile import ZipFile
from urllib.request import urlopen
from io import BytesIO
from bs4 import BeautifulSoup

wordFile = urlopen("http://pythonscraping.com/pages/AWordDocument.docx").read()
wordFile = BytesIO(wordFile)
document = ZipFile(wordFile)
xml_content = document.read("word/document.xml")

wordObj = BeautifulSoup(xml_content.decode("utf-8"), "xml")
textStrings = wordObj.find_all("w:t")

for textElem in textStrings:
    print(textElem.text)
```

```python
textStrings = wordObj.find_all("w:t")

for textElem in textStrings:
    style = textElem.parent.parent.find("w:pStyle")
    if style is not None and style["w:val"] == "Title":
        print("Title is: {}".format(textElem.text))
    else:
        print(textElem.text)
```
